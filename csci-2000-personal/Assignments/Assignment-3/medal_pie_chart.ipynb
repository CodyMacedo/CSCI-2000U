```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gold_list = [46,38,29,24,13,11,11,8,8,7,107]
country_list = ['USA','CHN','GBR','RUS','KOR','GER','FRA','ITA','HUN',
                'AUS','OTHER']
```

```python
plt.figure(figsize=(6,6), dpi=100)
plt.rcParams['font.size'] = 10.0
plt.rcParams['font.weight'] = 'bold' 


plt.pie(gold_list, explode=None, labels= country_list,
    colors=('lightcoral','yellowgreen','gold',
            'lightskyblue','b', 'g', 'r', 'c', 'm', 'y', 'orange'),
    autopct='%d', pctdistance=0.6, shadow=True,
    labeldistance=1.1, startangle=110,textprops=None)
plt.axis('equal')
plt.tight_layout()

plt.title('2012 Olympic gold medals (100486136)', 
           fontdict={'fontsize':20})


plt.show()
```
